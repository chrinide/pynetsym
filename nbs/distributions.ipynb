```python
import random as r; from pynetsym import mathutil
```

```python
ccdf = mathutil.ccdf
```

```python
def plot_ln(mu, sigma, points=10000):    points = [r.lognormvariate(mu, sigma) for _p in xrange(points)]    binvalues, ranges = histogram(points,bins=100)    return loglog(ranges[:-1], ccdf(binvalues))
```

```python
plot_ln(1,1)
```

```python
plot_ln(1,0.1)
```

```python
plot_ln(1,10)
```

```python
plot_ln(2,1)
```

```python
plot_ln(10,1)
```

```python
plot_ln(10,10)
```
